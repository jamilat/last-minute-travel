```python
data_path = "../data/obsolete/items_data.csv"
data_path_mod = "../data/obsolete/items_data_utf.csv"
```

```python
with open(data_path, "rb") as f:
    lines = [l.decode('utf8', 'ignore') for l in f.readlines()]

with open(data_path_mod, "w") as f:
    for line in lines:
        f.write(line)
```

```python
df_items_list = pd.read_csv(data_path_mod, na_filter=False)
df_items_list.drop(index=0, inplace=True)
```

```python
activities_items_dict = defaultdict(lambda: defaultdict(list))

for column in df_items_list.columns:
    lis = list(df_items_list[column])
    i = 1
    category = lis[0]
    while i < len(lis) - 1:
        if lis[i] == "" and lis[i+1] == "":
            break
        if lis[i] == "":
            category = lis[i+1]
            item_list = list()
            i += 2
        activities_items_dict[column][category].append(lis[i])
        i += 1
```

```python
df_activities_items = pd.DataFrame(columns=["activity", "category", "item"])

for activity, item_dict in activities_items_dict.items():
    for item_category, items in item_dict.items():
        df_activities_items.loc[len(df_activities_items)] = [activity, item_category.strip().lower(), [item.strip(" *").lower() for item in items]]
```

```python
df_activities_items.to_csv("../data/activities_items.csv", index=0)
```
